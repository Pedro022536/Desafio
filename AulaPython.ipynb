```python
#remover valores nulos
df = df.dropna()
```

```python
#remover coluna lunch
df.drop('lunch', axis=1, inplace=True)
```

```python
df
```

```python
#Média aritimetica
mediamat = df['math score'].mean()
print("A media das notas de matematica é:", mediamat)
```

```python
mediarede = df['reading score'].mean()
print("A media das notas de leitura é:", mediarede)
```

```python
mediawrit = df['writing score'].mean()
print("A media das notas de escrita é:", mediawrit)
```

```python
minmat = df['math score'].min()
print("A menor nota de matematica é:", minmat)
```

```python
maxmat = df['math score'].max()
print("A maior nota de matematica é:", maxmat)
```

```python
df.describe()
```

```python
df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2.plot(kind='bar')
```

```python
df2 = df.groupby('parental level of education')[['reading score']].mean()
df2.plot(kind='bar')
```

```python
sns.pairplot(df[['math score','reading score','writing score']], diag_kind='kde')
plt.show()
```

-Pairplot(ou gráfico de pares), muito utilizado para analisar a relação entre variáveis númericas em um conjunto de dados. Ele mostra

```python
sns.histplot(df['math score'], bins =20, kde=True)
plt.show()
```

```python
#contar frequência de cada categoria
race_count = df['race/ethnicity'].value_counts()
plt.pie(race_count, labels=race_count.index, autopct='%1.1f%%')
plt.title('Distribuição por raça/etnia')
plt.axis ('equal')
plt.show()
```
